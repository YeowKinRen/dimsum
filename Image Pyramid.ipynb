## Image Pyramid

1. Gaussian Pyramid 
- Used to downsample images
- formed by removing consecutive rows and columns in Lower level (higher resolution) image
- each pixel in higher level is formed by the contribution from 5 pixels in underlying level with gaussian weights
- area reduces to one-fourth of original area *Octave
2. Laplacian Pyramid
- Used to reconstruct an upsampled image from an image lower in the pyramid (with less resolution)
- formed from the Gaussian Pyramids
- edge images only, most of its elements are zeros.

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("images/Lenna.png")
```

`pyrUp()` function increases the size to double of its original size
`pyrDown()` function decreases the size to half

```python
layer = img.copy() 
  
for i in range(4): 
    plt.subplot(2, 2, i + 1) 
    layer = cv2.pyrDown(layer) 
    imgCvt = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB) 
    # convertion of OCV image that are represented in BGR
    plt.imshow(imgCvt)
    cv2.imshow("str(i)", layer) 
    cv2.waitKey(0) 
      
  
cv2.destroyAllWindows() 
```
